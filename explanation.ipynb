# Understanding My Project From Scratch

## What is this project in plain English?

Imagine I work at a telecom company like Airtel or Jio. Every month, some customers cancel their subscription and leave — this is called **churn**. Losing customers costs the company money.

My job is to build a system that looks at a customer's profile and answers one question:

> **"Is this customer likely to leave us next month?"**

That's exactly what this project does. It's like a **crystal ball for customer behaviour** — built using real data and machine learning.

---

## The Story of My Project (Step by Step)

### 📦 Step 1 — I got real data
I used the **IBM Telco Customer Churn dataset** — a real-world dataset with **7,043 actual customers** and 21 facts about each one, like:
- How long have they been a customer? (tenure)
- How much do they pay monthly?
- Do they have a phone plan? Internet? Tech support?
- Did they actually leave? (the answer you're trying to predict)

---

### 🧹 Step 2 — I cleaned and improved the data
Raw data is messy. I fixed problems like:
- **11 rows** had a blank space instead of a number in `TotalCharges` — I filled those in automatically
- I removed duplicate records
- I converted text like "Yes/No" into numbers the computer understands (1/0)

Then I **created 6 brand new features** that didn't exist in the original data but make the model smarter:

| Feature I invented | What it means |
|---|---|
| `AvgCharges` | How much does this customer pay per month on average? |
| `TenureGroup` | Are they a new, growing, loyal, or champion customer? |
| `ServiceCount` | How many extra services have they subscribed to? |
| `ChargeRatio` | Are they a new customer paying a lot? (high churn risk) |
| `HasHighValuePlan` | Are they on a long-term contract? (less likely to leave) |
| `IsSeniorAlone` | Is this a senior citizen with no family? (higher risk group) |

---

### 🤖 Step 3 — I trained three AI models
I taught three different algorithms to recognise patterns of churning customers:

1. **Logistic Regression** — the simple, fast baseline model
2. **Random Forest** — builds 200 decision trees and takes a vote
3. **XGBoost** — the most powerful, award-winning algorithm used by top data scientists worldwide

Each model looks at a customer's data and learns: *"customers with short tenure + fiber optic + month-to-month contract tend to leave."*

I also handled a big problem — **only 26.5% of customers churned**, so the data was imbalanced. I used a technique called **SMOTE** (Synthetic Minority Oversampling) to artificially create more churn examples so the model doesn't just always predict "won't churn."

---

### ✅ Step 4 — I tested the models fairly
I kept 20% of data hidden during training as a **test set** — data the model had never seen. Then I measured:

| Metric | What it means | My Score |
|---|---|---|
| **AUC** | How well does it rank churners above non-churners? | 0.871 |
| **F1 Score** | Balance of catching churners vs false alarms | 0.649 |
| **Accuracy** | Overall correct predictions | 82.3% |

An AUC of **0.871** means the model correctly identifies the at-risk customer **87% of the time** — far better than random guessing (50%).

---

### 🔍 Step 5 — I explained WHY the model decides what it decides
Using a technique called **SHAP** (SHapley Additive exPlanations), I can see exactly which factors pushed a prediction toward churn or away from it. For example:

> *"This customer is predicted to churn because: short tenure (+0.38), month-to-month contract (+0.31), fiber optic internet (+0.26)"*

This is called **Explainable AI** — crucial for real business use so managers can actually trust and act on the predictions.

---

### 🖥️ Step 6 — I built a live web dashboard
I built a professional 4-page web app using **Streamlit + Plotly**:
- **Analytics page** — 10 interactive charts exploring the data
- **Predict page** — enter any customer's details, get an instant churn probability with a risk gauge
- **Model Performance page** — ROC curve, confusion matrix, model comparison
- **Explainability page** — SHAP charts showing why the model predicted what it did
